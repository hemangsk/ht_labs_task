```python
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
df3 = pd.DataFrame(injured_data_2015)
df3 = df3.sort_values(by='States/UTs', axis=0)
plotted = df3.set_index('States/UTs').plot.bar(figsize=(15,10))
plt.show()
```
